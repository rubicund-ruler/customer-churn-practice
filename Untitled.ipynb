```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_original = pd.read_csv('Churn_Modeling.csv')
df_original = df_original.drop(columns=['CustomerId', 'Surname'])
df_original
```

```python
df = pd.get_dummies(df_original, prefix=['Geography', 'Gender'], dtype=int).drop(columns=["RowNumber"])
df
```

```python
# Making sure our dataset doesn't have nulls in important so we will count how many nulls are there in total
count_null = df.isnull().sum().sum()
print("Number of NaNs in dataset:", count_null)
```
