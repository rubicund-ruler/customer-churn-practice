```python
###### from statsmodels.stats.outliers_influence import variance_inflation_factor

# VIF dataframe
df_for_vif = df
df_for_vif = df_for_vif.drop(columns=["Exited"])
vif_data = pd.DataFrame()
vif_data["feature"] = df_for_vif.columns


# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(df_for_vif.values, i)
                          for i in range(len(df_for_vif.columns))]

print(vif_data)
```

```python
#Interactions
df['Age_Balance_Interaction'] = df['Age'] * df['Balance']
df['Germany_Balance_Interaction'] = df['Geography_Germany'] * df['Balance']
df
```

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop('Exited', axis=1)
y = df['Exited']

# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)
```

```python
from sklearn.model_selection import train_test_split
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the model
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Confusion matrix
print(confusion_matrix(y_test, y_pred))

# Classification report
print(classification_report(y_test, y_pred))
```
